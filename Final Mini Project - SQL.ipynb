### Tasks

***Find the number of books released after January 1, 2000.***

```python
pd.io.sql.read_sql(sql = """SELECT COUNT(*) AS "Number of Books" FROM books WHERE publication_date > '2000-01-01'""", \
                   con = engine)
```

`819 books were released after January 1, 2000.`

***Find the number of user reviews and the average rating for each book.***

```python
pd.io.sql.read_sql(sql = """SELECT b.book_id, b.title, COUNT(rs.*) AS "Number of User Reviews", \
                            AVG(r.rating) AS "Average Rating" FROM books AS "b" JOIN ratings AS "r" ON b.book_id = r.book_id \
                            JOIN reviews AS "rs" ON r.book_id = rs.book_id GROUP BY b.book_id, b.title \
                            ORDER BY "Average Rating" DESC""", con = engine)
```

`We see that a book can have the maximum rating and still have the lowest number of user reviews. For instance, A woman of substance has the highest average rating but a somewhat low number of reviews.`

***Identify the publisher that has released the greatest number of books with more than 50 pages (this will help you exclude brochures and similar publications from your analysis).***

```python
pd.io.sql.read_sql(sql = """SELECT p.publisher_id, p.publisher, COUNT(b.*) AS "Number of books" FROM books AS "b" \
                            JOIN publishers AS "p" ON b.publisher_id = p.publisher_id WHERE b.num_pages > 50 \
                            GROUP BY p.publisher_id, p.publisher ORDER BY "Number of books" DESC LIMIT 1;""", con = engine)
```

`Penguin books looks like the publisher that has the greatest nummber of books released.`

***Identify the author with the highest average book rating (look only at books with at least 50 ratings).***

```python
pd.io.sql.read_sql(sql = """SELECT a.author, b.book_id, AVG(r.rating) AS "Average book rating" FROM books AS "b" \
                            JOIN authors AS "a" ON b.author_id = a.author_id JOIN ratings AS "r" ON b.book_id = r.book_id \
                            GROUP BY b.book_id, a.author HAVING COUNT(r.rating) >= 50 ORDER BY "Average book rating" DESC \
                            LIMIT 1;""", con = engine)
```

`J.K. Rowling/Mary GrandPrÃ© top in this category with a 4.4 average book rating.`

***Find the average number of text reviews among users who rated more than 50 books.***

```python
pd.io.sql.read_sql(sql = """WITH reviewers AS (SELECT username, COUNT(*) AS "Number of reviews" FROM reviews \
                            WHERE username IN (SELECT username FROM ratings GROUP BY username HAVING COUNT(rating)> 50) \
                            GROUP BY username) SELECT username, AVG("Number of reviews") OVER() AS "Average number of text \
                            reviews" FROM reviewers;""", con = engine)
```

`24.3 is the average number of text reviews among users who rated more than 50 books.`
